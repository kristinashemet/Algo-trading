Create a column which calculates the differences between the opening and closing prices per day

```python
# Add a column `diff` to `aapl` 
aapl['diff'] = aapl.Open - aapl.Close

aapl.tail()
```

```python
# Plot the closing prices for `aapl`
aapl['Close'].plot(grid=True)

plt.show()
```

```python
# Plot the spread between open and close per day for `aapl`
aapl['diff'].plot(grid=True)

plt.show()
```

Calculate daily percentage change in the value of stock over a single day of trading (not taking into account dividends and other factors)

```python
# Assign `Adj Close` to `daily_close`
daily_close = aapl[['Adj Close']]

# Daily returns
daily_pct_c = daily_close.pct_change()

# Replace NA values with 0
daily_pct_c.fillna(0, inplace=True)

# Daily log returns
daily_log_returns = np.log(daily_close.pct_change()+1)

# Print daily log returns
print(daily_log_returns)
```

```python
# Plot the distribution of `daily_pct_c`
daily_pct_c.hist(bins=50)

plt.show()

# Pull up summary statistics
print(daily_pct_c.describe())
```

We can also calculate the cumulative daily returns

```python
# Calculate the cumulative daily returns
cum_daily_return = (1 + daily_pct_c).cumprod()

# Print `cum_daily_return`
print(cum_daily_return)
```

```python
# Plot the cumulative daily returns
cum_daily_return.plot(figsize=(12,8))

# Show the plot
plt.show()
```

Get daily data from yahoo finance of Apple, Google, IBM and Microsoft and concatenate in one dataset

```python
def get(tickers, startdate, enddate):
    def data(ticker):
        return (pdr.get_data_yahoo(ticker, start=startdate, end=enddate))
    datas = map (data, tickers)
    return(pd.concat(datas, keys=tickers, names=['Ticker', 'Date']))

tickers = ['AAPL', 'MSFT', 'IBM', 'GOOG']
all_data = get(tickers, datetime(2006, 10, 1), datetime(2023, 3, 3))
all_data.head()
```

```python
import pandas as pd 
all_data.to_csv('all_data.csv') 
df = pd.read_csv('all_data.csv', header=0, index_col='Date', parse_dates=True) 
```

```python
daily_close_px = all_data[['Adj Close']].reset_index().pivot('Date', 'Ticker', 'Adj Close')

# Calculate the daily percentage change for `daily_close_px`
daily_pct_change = daily_close_px.pct_change()

# Plot the distributions
daily_pct_change.hist(bins=50, sharex=True, figsize=(12,8))

# Show the resulting plot
plt.show()
```

```python
# Plot a scatter matrix with the `daily_pct_change` data 
pd.plotting.scatter_matrix(daily_pct_change, diagonal='kde', alpha=0.1,figsize=(12,12))

# Show the plot
plt.show()
```

```python
# Calculate the cumulative daily returns
cumul_daily_return = (1 + daily_pct_change).cumprod()

# Print `cum_daily_return`
print(cumul_daily_return)
```

```python
# Plot the cumulative daily returns
cumul_daily_return.plot(figsize=(12,8))

# Show the plot
plt.show()
```
